```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
```

```python
trueProbs = []
for i in range(39):
    trueProbs.append(np.load('fiberProbs_'+str(i)+'.npy'))
trueProbs = np.array(trueProbs)[:,:,0]*100
```

```python
plt.figure()
plt.scatter(np.arange(39),np.sum(trueProbs,axis=-1))
plt.ylim([0,40])
```
